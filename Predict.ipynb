```python
import pandas as pd

data = pd.read_csv('./voyages.csv')
data.reset_index(drop=True)
```

```python
mode = data.groupby('begin_port_id')['end_port_id'].agg(lambda x:x.value_counts().index[0])
mode[138]
```

```python
vessel = data['vessel']
output = pd.DataFrame(columns=['vessel', 'begin_port_id', 'end_port_id', 'voyage'])
for i in vessel:
  row = data.loc[data['vessel'] == i].iloc[-1]
  v1 = mode[row['end_port_id']]
  v2 = mode[v1]
  v3 = mode[v2]
  o = [i, row['begin_port_id'], row['end_port_id'], [v1, v2, v3]]
  output.loc[i] = o
```

```python
output.to_csv('predict.csv', index=False)
```
